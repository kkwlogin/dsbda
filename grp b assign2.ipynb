```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
air = pd.read_csv("C:/Users/HP/OneDrive/Desktop/DSBDA/Grp B Assign 2/airquality.csv") 
heart = pd.read_csv("C:/Users/HP/OneDrive/Desktop/DSBDA/Grp B Assign 2/heartdisease.csv")
```

```python
print(air.info())
print(heart.info())
```

```python
# Fill missing numerical values with mean
air.fillna(air.mean(numeric_only=True), inplace=True)
heart.fillna(heart.mean(numeric_only=True), inplace=True)

# Optional: drop duplicates
air.drop_duplicates(inplace=True)
heart.drop_duplicates(inplace=True)
```

```python
# Step 1: Display first 5 rows of cleaned air quality data
print("Cleaned Air Quality Data:")
display(air.head())

# Step 2: Display number of missing values after cleaning
print("\nMissing Values in Air Quality Dataset (After Cleaning):")
print(air.isnull().sum())

# Step 3: Display first 5 rows of cleaned heart disease data
print("\nCleaned Heart Disease Data:")
display(heart.head())

# Step 4: Display number of missing values after cleaning
print("\nMissing Values in Heart Disease Dataset (After Cleaning):")
print(heart.isnull().sum())

# Step 5: Display shape after dropping duplicates
print(f"\nShape after cleaning - Air: {air.shape}, Heart: {heart.shape}")
```

```python
# If no common key, just add side-by-side (concatenation)
combined = pd.concat([air.reset_index(drop=True), heart.reset_index(drop=True)], axis=1)

# Display combined data
combined.head()
```

```python
# Convert categorical to numerical if needed (example)
combined = pd.get_dummies(combined, drop_first=True)

# Normalize numerical features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(combined)

# Convert back to DataFrame
combined_scaled = pd.DataFrame(scaled_data, columns=combined.columns)
combined_scaled.head()
```

```python
# Convert categorical columns to numeric using one-hot encoding
combined_encoded = pd.get_dummies(combined, drop_first=True)

# Display the transformed columns
print("Encoded columns:")
display(combined_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Scale the data
scaler = StandardScaler()
combined_scaled_array = scaler.fit_transform(combined_encoded)

# Convert back to DataFrame
combined_scaled = pd.DataFrame(combined_scaled_array, columns=combined_encoded.columns)

# Display normalized data
print("Normalized Data:")
display(combined_scaled.head())
```

```python
# Remove irrelevant column
heart = heart.drop(columns=['Unnamed: 0'])

# Drop rows with missing or NaN values
heart = heart.dropna()

# Display cleaned shape
print("Cleaned heart dataset shape:", heart.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode target column (Yes=1, No=0)
heart['AHD'] = LabelEncoder().fit_transform(heart['AHD'])

# Convert categorical columns to dummy variables
heart_encoded = pd.get_dummies(heart, drop_first=True)

# Split features and target
X = heart_encoded.drop('AHD', axis=1)
y = heart_encoded['AHD']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Build logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("✅ Model Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
