```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/91992/OneDrive/Documents/sem-6/DSBD/practical/forestfire.csv')
```

```python
data.head()
```

### LINE PLOT

```python
# plot avg area burnbed month wise
data.groupby('month') ['area'].mean().plot(marker = 'o')
plt.xlabel("month")
plt.ylabel("avrage area burned")
plt.title("average area burned by month")
plt.grid(True)
plt.show()
```

```python
month_avg = data.groupby('month') ['area'].mean().reset_index()
sns.lineplot(x='month', y='area' , data=month_avg, color = 'skyblue')
plt.xlabel("month")
plt.ylabel("avrage area burned")
plt.title("average area burned by month")
plt.grid(True)
plt.show()
```

### bar plot

```python
# plot bar avg temp change month wise
month = data.groupby('month') ['temp'].mean().reset_index()
plt.figure(figsize = (6,4))
plt.bar(month['month'], month['temp'], color = 'skyblue')
plt.xlabel("month")
plt.ylabel("avrage temp")
plt.title("average temp by month")
plt.show()
```

```python
month = data.groupby('month') ['temp'].mean().reset_index()
plt.figure(figsize = (6,4))
sns.barplot(x='month', y='temp' , data=month, color = 'skyblue')
sns.set(style="whitegrid")
plt.xlabel("month")
plt.ylabel("avrage temp")
plt.title("average temp by month")
plt.show()
```

### histogram

```python
plt.hist(data['temp'], bins=10, color='coral', edgecolor='black')
plt.xlabel('temp')
plt.ylabel('frequency')
plt.show()
```

```python
sns.histplot(data['temp'], bins=10, color='coral', edgecolor='black')
plt.xlabel('temp')
plt.ylabel('frequency')
plt.show()
```

### scatter plot

```python
plt.scatter(data['temp'], data['area'], color = 'green')
plt.xlabel('temp')
plt.ylabel('area')
plt.show()
```

### pie chat

```python
day_count = data['day'].value_counts()
plt.pie(day_count, labels = day_count.index, autopct = '%1.1f%%', startangle = 90)
plt.axis('equal')
plt.show()
```

### box plot

```python
data.boxplot(column='temp', by='month')
plt.xlabel('month')
plt.ylabel('area')
plt.show()
```
